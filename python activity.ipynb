```python
#s1a.py
# Part A: Calculate months to save for down payment without raises
# Get user inputs
annual_salary = float(input("Enter annual salary: "))
portion_saved = float(input("Enter the percent of salary to save, as a decimal: "))
total_cost = float(input("Enter the cost of dream home: "))

# Calculate down payment and monthly values
down_payment = 0.25 * total_cost
monthly_salary = annual_salary / 12
monthly_saved = monthly_salary * portion_saved
annual_return = 0.04
monthly_return = annual_return / 12

# Initialize savings and months
current_savings = 0.0
months = 0


while current_savings < down_payment:
    # Apply monthly investment return
    current_savings += current_savings * monthly_return
   #add montky savings
    current_savings += monthly_saved
    # Increment months
    months += 1

print("Number of months:", months)
```

```python
# ps1b.py
# Part B: Calculate months to save for down payment with semi-annual raises

starting_annual_salary = float(input("Enter your starting annual salary: "))
portion_saved = float(input("Enter the percent of your salary to save, as a decimal: "))
total_cost = float(input("Enter the cost of your dream home: "))
semi_annual_raise = float(input("Enter the semi-annual raise, as a decimal: "))

# Calculate down payment and initial monthly values
down_payment = 0.25 * total_cost
annual_salary = starting_annual_salary
monthly_salary = annual_salary / 12
monthly_saved = monthly_salary * portion_saved
annual_return = 0.04
monthly_return = annual_return / 12

# Initialize savings and months
current_savings = 0.0
months = 0

# Simulate until down payment is reached
while current_savings < down_payment:
    # Check for semi-annual raise (every 6 months, starting after month 6)
    if months % 6 == 0 and months > 0:
        annual_salary *= (1 + semi_annual_raise)
        monthly_salary = annual_salary / 12
        monthly_saved = monthly_salary * portion_saved
    
    # Apply monthly investment return
    current_savings += current_savings * monthly_return
    # Add monthly savings
    current_savings += monthly_saved
    # Increment months
    months += 1


print("Number of months:", months)
```
